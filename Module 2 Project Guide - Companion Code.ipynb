### Querying price of cryptocurrencies using CoinGecko

Coingecko documentation: https://docs.coingecko.com/demo/reference/endpoint-overview

```python
url = "https://api.coingecko.com/api/v3/simple/price?vs_currencies={vs_currencies}&ids={ids}&x_cg_demo_api_key={API_KEY}"

params = {
    "vs_currencies": "usd",
    "ids": "bitcoin",
    "x_cg_demo_api_key": API_KEY,
}

r = requests.get(url, params=params)

# Prints API status code. If successful, we should see 200 OK.
print(r.status_code)

# Prints entire JSON response
print(r.json())
```

```python

API_KEY = "CG-iPH2VqFm7Ekk8o37G5U7Eymv"

url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency={vs_currency}&ids={ids}&x_cg_demo_api_key={API_KEY}"

params = {
    "vs_currency": "usd",
    "ids": "bitcoin",
    "names": "Bitcoin",
    "x_cg_demo_api_key": API_KEY
}

r = requests.get(url, params=params)

# Prints API status code. If successful, we should see 200 OK.
print(r.status_code)

# Prints entire JSON response in a nice way
print(json.dumps(r.json()[0], indent=2))
```

### Query using API keys stored in a separate env file (and not writen in the codebase)

1. Make sure you have already run "pip install python-dotenv" in the terminal
2. Create a new file in your project folder and call it .env with your .ipynb file in the same folder (VERY IMPORTANT or it would not work)
3. In the env, enter your API key (eg. COINGECKO_API_KEY=your_actual_api_key_here) -- do not includes quotation marks like "" 

```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key_stored = os.getenv("COINGECKO_API_KEY")

# Test if the API key stored in the .env file can be successfully queried
print(api_key_stored)

# Check which working directory you are in. If your python/ipynb file is not in this working directory and the env file is in this directory, it will fail to extract the key.
# os.getcwd()
```

```python
url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency={vs_currency}&ids={ids}&x_cg_demo_api_key={api_key_stored}"

params = {
    "vs_currency": "usd",
    "ids": "bitcoin",
    "x_cg_demo_api_key": api_key_stored
}

r = requests.get(url, params=params)

# Prints API status code. If successful, we should see 200 OK.
print(r.status_code)

# Prints entire JSON response in a nice way
print(json.dumps(r.json()[0], indent=2))

```
